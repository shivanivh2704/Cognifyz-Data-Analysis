```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv(r"C:\Users\dazvi\OneDrive\Desktop\cognifyz internship\Dataset .csv")
df.head()
```

```python
# Find the restaurant with the highest number of votes
highest_vote = df[df['Votes'] == df['Votes'].max()]
highest_vote
```

```python
highest_vote[['Restaurant Name', 'Votes']]
```

```python
# Find the restaurant with the lowest number of votes
lowest_vote = df[df['Votes'] == df['Votes'].min()]
lowest_vote[['Restaurant Name', 'Votes']]
```

```python
# Calculate the correlation between number of votes and rating
correlation = df['Votes'].corr(df['Aggregate rating'])

print("\nCorrelation Between Number of Votes and Rating:", correlation)
```

```python
# Create a scatter plot to analyze the correlation

plt.figure(figsize=(10, 6))
plt.scatter(df['Votes'], df['Aggregate rating'], alpha=0.5, color='cyan')
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate Rating')
plt.title('Correlation Between Votes and Rating')
plt.grid(axis='both', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
plt.figure.figsize=(5,3)

plt.bar(df['Votes'], df['Aggregate rating'])
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate Rating')
plt.title('Correlation Between Votes and Rating')
plt.show()
     
```

```python
# Create a bar chart to visualize the relationship with rotated X-axis labels
plt.figure(figsize=(10, 6))
plt.bar(df['Votes'], df['Aggregate rating'],  color='cyan')
plt.xlabel('Number of Votes')
plt.ylabel('Aggregate Rating')
plt.title('Correlation Between Votes and Rating')
plt.grid(axis='both', linestyle='--', alpha=0.7)


# Adjust the layout to prevent overlapping labels
plt.tight_layout()

plt.show()
```
